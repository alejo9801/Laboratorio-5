```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
path="C:/Users/Usuario/Documents/UVG 9no SEMESTRE\MODELACION Y SIMULACION"
file="ANALISIS DE DATOS/operacionesmayo.csv"
pathdata=os.path.join(path,file)
```

```python
#LECTURA CSV
data=pd.read_csv(pathdata)
#IMPORTACION DEL CSV
data.head(16)
```

```python
mostrarmetasc=data.loc[data['MetaCumplida']=='Yes',['TiendasVentas']]
mostrarmetasc.head(16)
```

```python
#GRAFICA DE VENTAS 2020
```

```python
vent=data['Ventas2020']
tien=data['TiendasVentas']
plt.rcParams["figure.figsize"] = (15, 15)
plt.pie(vent,labels=tien, shadow=True, autopct="%1.2f%%")
#plt.legend()
plt.show()
```

```python
#GRAFICA DE VENTAS 2020
```

```python
vent=data['Ventas2021']
tien=data['TiendasVentas']
plt.rcParams["figure.figsize"] = (15, 15)
plt.pie(vent,labels=tien, shadow=True, autopct="%1.2f%%")
#plt.legend()
plt.show()
```

```python
#SUMA DE VENTAS 2020 Y 2021
datossumar=data['Ventas2020']+data['Ventas2021']
data= data.assign(Total=datossumar.values)
#REMPLAZO DE YES =1 Y NO=0
data['MetaCumplida']=data['MetaCumplida'].replace('Yes','1')
data['MetaCumplida']=data['MetaCumplida'].replace('No','0')
data.head(16)
```

```python
#Guardar NUEVO EXCEL
NUEVO_ACRCHIVO_1="/ANALISIS DE DATOS."
data.to_excel(path+"/"+NUEVO_ACRCHIVO_1+"xlsx")
```
